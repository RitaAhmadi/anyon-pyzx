```python
import pyzx as zx
import numpy as np
from fractions import Fraction
```

```python
class Anyons:
    global phi, theta 
    phi = (1+np.sqrt(5))/2
    theta = 2 * np.arccos(1/phi)

    
    def __init__(self, name):
        self.name = name
    
    
    def b3(name):
        if name=="isg":
            R1 = zx.Circuit(1)
            R1.add_gate("ZPhase", 0, Fraction(-1, 2))
            
            R2= zx.Circuit(1)
            R2.add_gate("XPhase", 0, Fraction(1, 2))
            return (R1, R2)
            
        elif name=="fib":
            R1 = zx.Circuit(1)
            R1.add_gate("ZPhase", 0, 1)
            R1.add_gate("ZPhase", 0, Fraction(2, 5))
            
            
            R2 = zx.Circuit(1)
            R2.add_gate("ZPhase", 0,Fraction(1, 2))
            R2.add_gate("XPhase", 0, theta)
            R2.add_gate("ZPhase", 0, Fraction(2, 5))
            R2.add_gate("XPhase", 0, theta)
            R2.add_gate("ZPhase", 0, Fraction(1, 2))
            
            return(R1, R2)
            
    def b6(name):
        if name=="isg":
            R1= zx.Circuit(2)
            R1.add_gate("ZPhase", 0, Fraction(-1, 2))
            
            R2 = zx.Circuit(2)
            R2.add_gate("XPhase", 0, Fraction(1, 2) )
            
            R3 = zx.Circuit(2)
            R3.add_gate("CNOT", 0, 1)
            R3.add_gate("XPhase", 0, Fraction(-1, 2))
            R3.add_gate("CNOT", 0, 1)
            
            R4= zx.Circuit(2)
            R4.add_gate("XPhase", 1, Fraction(1, 2))
            
            R5= zx.Circuit(2)
            R5.add_gate("ZPhase", 1, Fraction(-1, 2))
            
            return (R1, R2, R3, R4, R5)
        
        
        elif name=="fib":
            U1 = zx.Circuit(3)
            U1.add_gate("XPhase", 0, 1)
            U1.add_gate("XPhase", 0, Fraction(7, 5))
            U1.add_gate("XPhase", 0, 1)
            
             
            U2 = zx.Circuit(3)
            U2.add_gate("CNOT", 0, 1)
            U2.add_gate("ZPhase", 1, Fraction(-7, 10))
            U2.add_gate("CNOT", 0, 1)
            U2.add_gate("ZPhase", 0, Fraction(7, 5))
            U2.add_gate("ZPhase", 0, Fraction(7, 10))
            
            U3 = zx.Circuit(3)
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(1, 4))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(-1, 4))
            U3.add_gate("HAD", 1)
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("HAD", 1)
            U3.add_gate("XPhase", 1, Fraction(-theta, 2))
            U3.add_gate("HAD", 1)
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("HAD", 1)
            U3.add_gate("XPhase", 1, Fraction(theta, 2))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(1, 4))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(-1, 4))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(-7, 10))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 0, Fraction(7, 5))
            U3.add_gate("ZPhase", 1, Fraction(-7, 10))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(-1, 4))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(1, 4))
            U3.add_gate("CZ", 0, 1)
            U3.add_gate("XPhase", 1, Fraction(theta, 2))
            U3.add_gate("CZ", 0, 1)
            U3.add_gate("XPhase", 1, Fraction(-theta, 2))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(-1, 4))
            U3.add_gate("CNOT", 0, 1)
            U3.add_gate("ZPhase", 1, Fraction(1, 4))
            
            U4 = zx.Circuit(3)
            U4.add_gate("XPhase", 1, 1)
            U4.add_gate("CNOT", 2, 1)
            U4.add_gate("ZPhase", 1, Fraction(-7, 10))
            U4.add_gate("CNOT", 2, 1)
            U4.add_gate("ZPhase", 1, Fraction(7, 10))
            U4.add_gate("ZPhase", 2, Fraction(7, 5))
            U4.add_gate("XPhase", 1, 1)
            
            U5 = zx.Circuit(3)
            U5.add_gate("CNOT", 1, 2)
            U5.add_gate("ZPhase", 2, Fraction(-7, 20))
            U5.add_gate("CNOT", 1, 2)
            U5.add_gate("ZPhase", 2, Fraction(7, 20))
            U5.add_gate("ZPhase", 1, Fraction(7, 10))
            U5.add_gate("CNOT", 0, 1)
            U5.add_gate("CNOT", 1, 2)
            U5.add_gate("ZPhase", 2, Fraction(7, 20))
            U5.add_gate("CNOT", 1, 2)
            U5.add_gate("ZPhase", 2, Fraction(-7, 20))
            U5.add_gate("ZPhase", 1, Fraction(-7, 10))
            U5.add_gate("CNOT", 0, 1)
            U5.add_gate("CNOT", 0, 2)
            U5.add_gate("ZPhase", 2, Fraction(7, 20))
            U5.add_gate("CNOT", 0, 2)
            U5.add_gate("ZPhase", 2, Fraction(7, 20))
            U5.add_gate("ZPhase", 0, Fraction(7, 10))
            
            
            U7= zx.Circuit(3)
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(-1, 8))
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(1, 8))
            U7.add_gate("ZPhase", 1, Fraction(1, 4))
            U7.add_gate("CNOT", 0, 1)
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(1, 8))
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(-1, 8))
            U7.add_gate("ZPhase", 1, Fraction(-1, 4))
            U7.add_gate("CNOT", 0, 1)
            U7.add_gate("CNOT", 0, 2)
            U7.add_gate("ZPhase", 2, Fraction(-1, 8))
            U7.add_gate("CNOT", 0, 2)
            U7.add_gate("ZPhase", 2, Fraction(1, 8))
            U7.add_gate("ZPhase", 0, Fraction(1, 4))
            U7.add_gate("HAD", 2)
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(-theta, 4))
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(theta, 4))
            U7.add_gate("ZPhase", 1, Fraction(theta, 2))
            U7.add_gate("CNOT", 0, 1)
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(theta, 4))
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(-theta, 4))
            U7.add_gate("ZPhase", 1, Fraction(-theta, 2))
            U7.add_gate("CNOT", 0, 1)
            U7.add_gate("CNOT", 0, 2)
            U7.add_gate("ZPhase", 2, Fraction(-theta, 4))
            U7.add_gate("CNOT", 0, 2)
            U7.add_gate("ZPhase", 2, Fraction(-theta, 4))
            U7.add_gate("ZPhase", 0, Fraction(theta, 2))
            U7.add_gate("HAD", 2)
            U7.add_gate("ZPhase", 2, Fraction(-theta, 8))
            U7.add_gate("CNOT", 0, 2)
            U7.add_gate("ZPhase", 2, Fraction(theta, 8))
            U7.add_gate("ZPhase", 0, Fraction(-theta, 4))
            U7.add_gate("CNOT", 0, 2)
            U7.add_gate("CNOT", 0, 1)
            U7.add_gate("ZPhase", 2, Fraction(theta, 8))
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(-theta, 8))
            U7.add_gate("ZPhase", 1, Fraction(theta, 4))
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("CNOT",0, 1)
            U7.add_gate("ZPhase", 2, Fraction(-theta, 8))
            U7.add_gate("CNOT", 1, 2)
            U7.add_gate("ZPhase", 2, Fraction(theta, 8))
            U7.add_gate("ZPhase", 1, Fraction(-theta, 4))
            U7.add_gate("CNOT", 1, 2)
        return(U1, U2, U3, U4, U5, U7)
```

```python
isg = Anyons.b3("isg")
fib = Anyons.b3("fib")
r1= isg[0]
r2=isg[1]
```

```python
lhs = zx.Circuit(1)
lhs = r1+r2+r1+r2.adjoint()+r1.adjoint()+r2.adjoint()
zx.draw(lhs)
g = lhs.to_graph()
```

```python
zx.full_reduce(g)
zx.draw(g)
```
